```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sn
```

```python
data = pd.read_csv("heart.csv")
```

```python
data.info()
```

```python
data.describe()
```

```python
data.duplicated().sum()
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
x= data["sex"]
y= data["age"]
```

```python
plt.plot(x,y,marker='o')
plt.xlabel("Sex",c='r')
plt.ylabel("Age",c="blue")
plt.title("Detail")
```

```python
plt.plot(x,y,ls=':')
plt.show()
```

```python
plt.plot(x,y,lw="2.9")
plt.show()
```

```python
plt.barh(x,y)
xfont= {"color":'red',"size":"2.5","family":'serif'}
plt.xlabel("Test",fontdict=xfont)
plt.grid(axis='x')
plt.show()
```

```python
plt.hist(y,x)
xfont= {"color":'red',"size":"2.5","family":'serif'}
plt.xlabel("Test",fontdict=xfont)
plt.grid(axis='x')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("heart.csv")
```

```python
data.head()
```

```python
data.sex.value_counts()
```

```python
data.value_counts("sex")
```

```python
sns.countplot(data=data,x="sex")
```

```python
NoMale = data[data.sex == 0]
NoFemale = data[data.sex == 1]
print(NoMale)
print(NoFemale)
```

```python
# under sampling
noFemaleS = NoFemale.sample(n = 91)
```

```python
noFemaleS
```

```python
newData = pd.concat([noFemaleS,NoMale])
```

```python
newData
```

```python
newData.sex.value_counts()
```

```python
sns.countplot(data= newData,x="sex")
```

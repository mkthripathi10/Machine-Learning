```python
df = sn.load_dataset("tips")
df.head()
```

```python
# scatter plot
sn.scatterplot(data=df ,x="total_bill",y="tip",hue="smoker",style="time")
plt.show()
```

```python
# Bar plot

sn.barplot(data=df,y="total_bill",x="day",hue="smoker")
plt.show()
```

```python
# histogram
sn.histplot(data=df,x="total_bill",hue="smoker")
```

```python
# boxplot
sn.boxplot(data=df,x="day",y="total_bill",hue="smoker")
```

```python
sn.violinplot(data=df,x="day",y="total_bill",hue="smoker")
```

```python
# pairplot

sn.pairplot(df,hue="smoker")
```

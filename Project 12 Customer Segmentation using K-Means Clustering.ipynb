```python
X =data[["Annual Income (k$)","Spending Score (1-100)"]].values
```

```python
print(X)
```

```python
wcss =[]
for i in range(1,11):
    MK = KMeans(n_clusters=i , init='k-means++',random_state=42)
    MK.fit(X)
    wcss.append(MK.inertia_)
```

```python

# sn.lineplot(x=range(1, 11), y=wcss)
# sn.scatterplot(x=range(1, 11), y=wcss)
plt.plot(range(1, 11), wcss)
plt.title("Elbow method for Optimal K")
plt.xlabel("No. of Cluster")
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans2 = KMeans(n_clusters=5 ,init="k-means++",random_state=42)
Y=kmeans2.fit_predict(X)
```

```python
print(Y)
```

```python
data["Cluster"]=Y
```
